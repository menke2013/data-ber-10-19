```python
def get_player_game_stats(playerid):
    
    url = f'https://stats.nba.com/stats/playerprofilev2?LeagueID=00&PerMode=Totals&PlayerID={playerid}'
    call1 = requests.get(url, headers= headers).json()
    call2 = call1['resultSets'].copy()
    call3 = call2[0].copy()
    headerlist = call3['headers']
    df_player = pd.DataFrame.from_dict(call3['rowSet'])
    if len(df_player) > 0:
        
        df_player.columns = headerlist
        df_player2 = df_player[::-1].copy()
        df_player2.reset_index(drop=True, inplace=True)
        df_current_season = df_player2.loc[df_player2['SEASON_ID'] == '2019-20']
    
        return df_current_season
        
    else: 
        return df_player



```

```python
df_stats_current = pd.DataFrame()

for playerid in list_playerID_current_season:
    print(len(df_stats_current))
    delay = random.randint(5,25)
    print(delay)
    time.sleep(delay)
    df_temp = get_player_game_stats(playerid)
#   print(df_temp)

    if len(df_temp) > 0:
    
        df_stats_current = df_stats_current.append(df_temp, ignore_index=True)


```
