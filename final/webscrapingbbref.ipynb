```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup

pd.set_option("display.max_columns",40)
```

```python
def dropping_header_in_table(df):
    drop_list_a = df.loc[df['Player'] == 'Player'].index.tolist()
    df.drop(drop_list_a, inplace=True)
    df.reset_index(drop=True)
    
    return df
```

```python
def changing_types(df):
    full_key_list = df.keys()
    
    sub_list_1 = ['Rk', 'Player', 'Pos', 'Age', 'Tm', 'G', 'MP', 'GS']
    list_integer = ['Rk', 'Age', 'G', 'MP']
    list_float = [x for x in full_key_list if x not in sub_list_1]
    
    
    for a in list_integer:
        df[a] = df[a].astype(int)
        
    for b in list_float:
        df[b] = df[b].astype(float)
        
    return df
        
    
```

```python
def cleaning_multiples(df):
    list_of_multiples = df['Player'].value_counts()[df['Player'].value_counts() != 1].keys().tolist()
    
    for l in list_of_multiples:
        drop_list_multiples = df.loc[(df['Player'] == l) & (df['Tm'] != 'TOT')].index.tolist()
        df.drop(drop_list_multiples, inplace=True)
        
    df.reset_index(drop=True)
    return df
    
```

```python
salaries_page = requests.get('https://www.basketball-reference.com/contracts/players.html')
```
