```python
def convert_imp_to_metric(x):
    
    ht = x.split('-')
    height = (int(ht[0])*12) + int(ht[1])
    height_cm = round(height *2.54, 1)
    
    return height_cm
    
 
    
    
```

```python
df_nba_stats_2019['HEIGHT'] = df_nba_stats_2019['HEIGHT'].apply(lambda x: convert_imp_to_metric(x))
```

```python
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=8, max_iter=300, random_state=42)
```

```python
df_2019_all5 = df_2019_all[['Player','Pos', 'FG%','3P%', '2P%', 'FT%', 'ORB', 'DRB', 'TRB','AST', 'STL', 'BLK', 'TOV', 'PTS', 'ORtg', 'DRtg', '3PAr', 'BPM', 'DBPM', 'DRB%', 'DWS', 'FTr', 'OBPM', 'ORB%', 'OWS',
       'PER', 'STL%', 'TOV%', 'TRB%', 'TS%', 'USG%', 'VORP', 'WS', 'WS/48','HEIGHT', 'WEIGHT']].copy()
```

```python
df_2019_all5
```

```python
columns_kmean = df_2019_all5._get_numeric_data().dropna(axis=1)
```

```python
kmeans_model.fit(columns_kmean)
labels = kmeans_model.labels_
```

```python
from sklearn.decomposition import PCA
pca_2 = PCA(2)
plot_columns = pca_2.fit_transform(columns_kmean)
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=labels)
plt.savefig('PlayerClustering')
plt.show()
```
