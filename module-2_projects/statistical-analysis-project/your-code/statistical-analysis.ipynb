```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
pd.set_option("display.max_columns",150)
```

```python
df1 = pd.read_csv('train.csv')
```

```python
df1.head(20)
```

```python
df1['SalePrice'].hist(log=True);
```

```python
df1.isna().sum()
```

```python
#we need some data cleaning here
```

```python
list1 = df1.keys()
list1
```

```python
list_mis_val_names = []
list_mis_val_percent =[]
for col in list1:    
    a = (df1[col].isna().sum())/(len(df1[col]))*100      
       
    if a > 0:
        list_mis_val_names.append(col)
        list_mis_val_percent.append(a)
    
```

```python
list_mis_val_names
```

```python
list_mis_val_percent
```

```python
#we are dropping all columns with more than 20 percent missing value
```

```python
list_dropna =[]
for i in range(len(list_mis_val_percent)):
    if list_mis_val_percent[i] > 20:
        list_dropna.append(list_mis_val_names[i])
        
```

```python
list_dropna
```

```python
df2 = df1.drop(list_dropna,axis=1).copy()
df2.head(20)
```

```python
#we are now dropping rows with missing values but we want to keep the rest of thse columns
# LotFrontage, the Garage columns, which are connected, basement columns and some masonery columns.  
# in the end we can keep 75 percent of our data and it is cleaned up from missing value. 
```

```python
df3 = df2.dropna().copy()
df3.head(20)
```

```python
df3.isna().sum().sum()
```

```python
list_drop2 = ['Street', 'Utilities', 'Condition2', 'RoofMatl', 'GarageCond', 'MiscVal', 'KitchenAbvGr', 'PavedDrive', 'LowQualFinSF', 'MSSubClass']

df3.drop(list_drop2, axis=1, inplace=True)
```

```python
#dropping some of the lot columns because the area will be the decisive factor and the others are quantifiable. 
# Also cleaning some garage stuff

list_drop3 = ['LotFrontage','LotShape', 'LandContour','LotConfig','LandSlope','GarageYrBlt',  'GarageFinish', 'GarageType', 'GarageCars']
df3.drop(list_drop3, axis=1, inplace=True)
```

```python
df3.reset_index(drop=True, inplace=True)
df3.head(20)
```

```python
#quality columns manipulation
list4 = df3.columns[df3.columns.str.contains('Q')]
df3[list4].head(20)
```

```python
list5 = list4.tolist().copy()
```

```python
list5.remove('OverallQual')
```

```python
list5
```

```python
dic_replace_qual = {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}
for i in list5:
    df3[i] = df3[i].map(dic_replace_qual)
```

```python
df3['Bath_nr'] = df3[df3.columns[df3.columns.str.contains('Bath')]].sum(1)
```
