```python
import pandas as pd
import numpy as np
import os
```

```python
#Importing the data from a csv file
not_a_dic = pd.read_csv('GSAF5.csv',sep=',', engine='python')
```

```python
not_a_dic.head()
```

```python
#Checking how much data is missing in every column in percent
not_a_dic.isna().sum()*100/len(not_a_dic)
```

```python
#copy for data cleaning and manipulation
not_a_dic2 = not_a_dic.copy()
```

```python
#optimizing the column names for easier working
not_a_dic2.columns = (not_a_dic2.columns
                      .str.lower()
                      .str.strip()
                      .str.replace(' ','_')
                      .str.replace('.','_')
                     )
not_a_dic2.columns
```

```python
#Checking for duplicated rows. All of them seem unique
not_a_dic2[not_a_dic2.duplicated()]
```

```python
#We are dropping every column with more than 40 percent missing data. If Age, Species or Time is importan we could subsetting a new table and work with half of the data. Decided to cut them al
drop_list = [l for l in not_a_dic2 if not_a_dic2[l].isna().sum()/len(not_a_dic2) > .4]
drop_list
not_a_dic2 = not_a_dic2.drop(drop_list, 1)
```

```python
not_a_dic2.shape
```

```python
#columns of Case Number, Case Number.1 and Case Number.2 are looking the same. Checking it to make sure. If yes we can drop 2 columns
not_a_dic2.query('case_number != case_number_1')
```

```python
not_a_dic2.query('case_number != case_number_2')
```

```python
#combining the query 
not_a_dic2.query('case_number != case_number_1 & case_number != case_number_2')
```

```python
#only 2 columns are different and those looking like typos. We can drop case_number_1 and case_number_2
not_a_dic2 = not_a_dic2.drop(['case_number_1', 'case_number_2'], 1)
```

```python
#Taking care of columns pdf, href and href_formula. They are all handling the reference
#pdf has no missing data, href_formula 1 and href 3
```

```python
not_a_dic2.loc[not_a_dic2['href_formula'].isna()]
```

```python
len(not_a_dic2.query('href_formula != href'))
#Difference is under 1 percent. 
```

```python
#We can use the href value to fill our href_value. With more than 1 row we would need to use a for loop.

not_a_dic2.loc[pd.isna(not_a_dic2['href_formula']), 'href_formula'] = not_a_dic2.loc[pd.isna(not_a_dic2['href_formula']),'href']
```

```python
not_a_dic2['href_formula'].isna().sum()
```

```python
#Now we can drop pdf and href column

not_a_dic2 = not_a_dic2.drop(['pdf','href'],1)
```

```python
not_a_dic2.isna().sum()*100/len(not_a_dic2)
```

```python
#We getting rid of some unreliable data by reducing the time frame

columns_keep = not_a_dic2.columns

not_a_dic3 = not_a_dic2[(not_a_dic2['year']>=1800)][columns_keep]
```

```python
#We also getting rid of the original order because we using a new index later. The date column is also too unspecific in it's values

not_a_dic3 = not_a_dic3.drop(['date','original_order'], 1)
```

```python
#cleaning up the fatal column
not_a_dic3['fatal_(y/n)'] = (not_a_dic3['fatal_(y/n)']
                             .str.strip()
                             .str.upper()
                             .str.replace('F', 'ND')
                             .str.replace('UNKNOWN','ND')
                             .str.replace('#VALUE!', 'ND')
                             .fillna('ND')
                    )
```

```python
#cleaning sex column
not_a_dic3['sex'] = (not_a_dic3['sex']
                             .str.strip()
                             .str.upper()
                             .str.replace('LLI', 'M')
                             .str.replace('N','ND')
                             .str.replace('.', 'ND')
                             .fillna('ND')
                    )
```

```python
#cleaning up the country column by dropping the NAN rows
not_a_dic3['country'].isna().sum()
```

```python
not_a_dic3.dropna(subset=['country'],inplace=True)
```

```python
#drop the name column because the information shouldn't be relevant unless for insurance reasons, but then it is filled with too much bad data
not_a_dic3 = not_a_dic3.drop('name',1)
```

```python
#filling the NAN in the other columns with ND (No Data)
not_a_dic3.fillna('ND', inplace=True)
```

```python
not_a_dic3.isna().sum()*100/len(not_a_dic3)
```

```python
#Renaming for better understanding
not_a_dic3.rename(columns ={'fatal_(y/n)': 'death',
                            'investigator_or_source': 'source',
                            'href_formula': 'reference',
                            'type' : 'type of accident',
                            'case_number' : 'case number'},
                  inplace=True)
```

```python
#Reseting the index
not_a_dic3.reset_index(drop=True, inplace=True)
```

```python
#writing into a pickle file
not_a_dic3.to_pickle('new_shark_in_town.pkl')
```
